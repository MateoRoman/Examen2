4. Transformamos nuestros datos categoricos a numericos


```python
# Asignar nombres a las columnas
column_names = ['age', 'workclass', 'fnlwgt', 'education', 'education_num',
                'marital_status', 'occupation', 'relationship', 'race', 'sex',
                'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'income']
data.columns = column_names
```

```python
# Imputar valores nulos con la media o la moda según sea apropiado
for column in data.columns:
    if data[column].isnull().sum() > 0:
        if data[column].dtype == 'object':
            data[column].fillna(data[column].mode()[0], inplace=True)
        else:
            data[column].fillna(data[column].mean(), inplace=True)
```

```python
# Codificar columnas categóricas
label_encoders = {}
categorical_columns = data.select_dtypes(include=['object']).columns

for col in categorical_columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le
```

```python
# Normalizar los datos
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
# Convertir de nuevo a DataFrame para facilitar el manejo
data_scaled = pd.DataFrame(data_scaled, columns=data.columns)
```

```python
# Guardar el DataFrame procesado en un nuevo CSV
processed_file_path = 'data_evaluacion_processed.csv'
data_scaled.to_csv(processed_file_path, index=False)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

# Cargar el dataset original
data_original = pd.read_csv('data_evaluacion.csv', header=None)

# Asignar nombres a las columnas del dataset original
data_original.columns = ["age", "workclass", "fnlwgt", "education", "education-num", "marital-status",
                         "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss",
                         "hours-per-week", "country", "salary"]

# Reemplazar los valores faltantes representados por '?' con NaN
data_original.replace('?', np.nan, inplace=True)

# Rellenar los valores faltantes con la moda (para columnas categóricas) y la media (para columnas numéricas)
for column in data_original.columns:
    if data_original[column].dtype == 'object':
        data_original[column].fillna(data_original[column].mode()[0], inplace=True)
    else:
        data_original[column].fillna(data_original[column].mean(), inplace=True)

# Convertir las columnas categóricas a numéricas usando LabelEncoder
label_encoders = {}
categorical_columns = ["workclass", "education", "marital-status", "occupation", "relationship", "race", "sex", "country", "salary"]

for column in categorical_columns:
    le = LabelEncoder()
    data_original[column] = le.fit_transform(data_original[column])
    label_encoders[column] = le

# Guardar el dataset limpio
data_original.to_csv('data_evaluacion_processed.csv', index=False)

print("El dataset ha sido limpiado y guardado como 'data_evaluacion_processed.csv'")
```
